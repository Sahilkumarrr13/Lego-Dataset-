# Importing datasets and libraries:

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./datasets/lego_sets.csv')
df.head()
```

```python
theme = pd.read_csv('./datasets/parent_themes.csv')
theme.head()
```

To start with let us merge both the dataframes into on dataframe.

For this we know that df has column parent_theme which is common with name in theme dataframe:

```python
df1 = df.merge(theme, left_on="parent_theme", right_on="name")
df1
```

Dropping the unecessary column and rename other columns:

```python
df1.drop("name_y", axis = 1, inplace = True)
df1.rename(columns ={"name_x":"name"}, inplace=True)
df1
```

## Q What percentage of all licensed sets ever released were Star Wars themed?

What we can do is filter our data for the only true licensed:

```python
df_licensed = df1[df1["is_licensed"]]
df_licensed
```

Let us perform some cleaning before solving the question:

```python
df_licensed.isnull().sum()
```

We need to drop the null values from set_num as given:

```python
df_licensed[df_licensed["set_num"].isnull()]
```

```python
df_licensed = df_licensed.dropna(subset=["set_num"])
df_licensed
```

```python
df_licensed.isna().sum()
```

Therefore are no more nan values in set_num.

Next we have missing values in num_parts but it is given that these are acceptable as they are not the center of analysis so we will leave them as it is

Now comming back to the question:

```python
star_wars = df_licensed[df_licensed["parent_theme"]=="Star Wars"]

the_force = int(star_wars.shape[0]/df_licensed.shape[0]*100)
the_force
```

## Q In which year was Star Wars not the most popular licensed theme (in terms of number of sets released that year)?

```python
df_licensed
```

```python
df_licensed.loc[:, "count"] = 1
df_licensed
```

```python
df_group = df_licensed.groupby(["year", "parent_theme"], as_index=False).sum()
max_df = df_group.sort_values(by = "count", ascending=False).drop_duplicates(["year"])
max_df
```

```python
new_era = max_df.iloc[0, 0]
new_era
```
